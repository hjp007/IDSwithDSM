```python
import numpy as np
import math
import matplotlib.pyplot as plt


```

```python
def gd(x, mu=0, sigma=1):
    """根据公式，由自变量x计算因变量的值

    Argument:
        x: array
            输入数据（自变量）
        mu: float
            均值
        sigma: float
            方差
    """
    left = 1 / (np.sqrt(2 * math.pi) * np.sqrt(sigma))
    right = np.exp(-(x - mu)**2 / (2 * sigma))
    return left * right
```

```python
x = np.arange(0, 1, 0.01)
#  因变量（不同均值或方差）
y_1 = gd(x, 0.5, 0.01)
plt.plot(x, y_1*100, color='green')

#  设置坐标系
# plt.xlim(0, 1)
# plt.ylim(-0.2, 5)
font2 = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 15,
}
plt.yticks([])

plt.xlabel("feature values after normalization", font2)
plt.ylabel('occurance', font2)
plt.savefig("1_occurance_example_1.eps",format="eps")
```
